```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
os.chdir('results')
all_files = os.listdir()
lidx_foi = np.where(["_km" in file for file in all_files])[0].tolist()
files_of_interest = np.array(all_files)[lidx_foi]
```

```python
lat = np.arange(-90, 105, 15)
n_lat = len(lat)

z = np.concatenate([np.arange(1,28,3), np.arange(30,75,5)],axis=0)
n_z = len(z)

res = np.zeros((n_z,n_lat))
```

```python
for _,file in enumerate(files_of_interest):
    data = np.loadtxt(file, skiprows=4)
    if "n_" in file:
        pos_lat = np.where(-int(file[3:5]) == lat)[0]
        pos_z = np.where(int(file[9:11]) == z)[0]
    else:
        pos_lat = np.where(int(file[3:5]) == lat)[0]
        pos_z = np.where(int(file[8:10]) == z)[0]
    res[pos_z, pos_lat] = data[data.shape[0]-1,data.shape[1]-2]

plt.figure(figsize=(10, 6))

plt.imshow(res, aspect='auto', cmap='viridis', origin='lower', interpolation='nearest')
plt.colorbar(label='Ozone concentration [cm^-3]')
plt.xticks(np.arange(len(lat)), lat)  # Latitude ticks
plt.yticks(np.arange(len(z)), z)  # Altitude ticks
plt.xlabel('Latitude (°)')
plt.ylabel('Altitude (m)')
plt.title('Colormap of Altitude vs Latitude')

plt.show()
```
